## 1.5. Creating Functions

```python
# Functions alow you to write a block of code given some parameters. It helps when you would repeat the same
# code multiple times and makes the code more organized.
# It is a good practice, when writing some complicated code, to first write parts as functions, test this parts,
# and then put the parts together by calling the function
# The format of the function is name_of_the_function(parameters_of_the_function)

# For example, the function "print_squares" below whites the squares of all numbers between a and b

def print_squares(a, b):
  for i in range(a, b):
    print(i**2)
```

```python
# Now we call the function so it prints all squares between 3 and 8:
print_squares(3,8)
```

```python
# The function below checks if the number is prime, by 
# checking if it is divisible by any other number, by checking 
# all numbers one by one (there are more efficient ways of making this)

# It returns True if is prime, and False if not prime
def is_prime(number):
  root = round(number ** 0.5) # checks all numbers up to its square root
  for n in range(2, root+1):
    if number % n == 0: # The operator % calculates the division remainder
      return False
  return True
```

```python
# We can test this function with numbers we are familiar with
print(is_prime(17)) # Prime
print(is_prime(27)) # Not Prime
print(is_prime(37)) # Prime
print(is_prime(47)) # Prime
print(is_prime(31247)) # Prime
```

```python
# We can create a function that calls the first function
# Lets write a function to print all the primes between two numbers

def find_primes(a, b):
  for n in range(a, b):
    if is_prime(n): # is_prime will return True or False
      print(n)
```

```python
# We can test our function with some bigger numbers
find_primes(2000, 2100) # Will print all the primes between 2000 and 2010
```

```python
# For example, the library numpy allows to make matrices operations, 
#such as create matrices, sum, multiply, transpose and invert

matrix_in_list = [[1,2,3],[3,2,1]]
A = np.array(matrix_in_list) # Transform the list into a 2 x 3 matrix
print("A: ","\n", A)

B = np.array(A) # Copies matrix A
C = np.ones((2,3)) # Create a 2 x 3 matrix os ones
print("C: ","\n", C)

D = (B+C).transpose() # Sum matrices B and C, and transpose
print("D: ","\n", D)

E = np.matmul(A,D) # Multiply matrices A (2 x 3) and D (3 x 2), obtaining a 2 x 2 matrix
print("E: \n", E)

print("E_invert: \n",np.linalg.inv(E)) # Invert the matrix
```

## 3.1. Flipping coins

```python
# Import libraries 
from numpy import random # Generate random numbers
from scipy.stats import norm # normal distribution
import statsmodels.api as sm # linear regression
```

```python
# Generate 10 random numbers between 0 and 1
# These are random numbers, so I get different numbers each time I run
random.random(10)
```

```python
# Flip 10 fair coins
# We generate random numbers: They are Head if the random number is < 0.5
# These are random, so we get a different number each time we run

coins = (random.random(10)) < 0.5 # "coin" will be True if Heads, False if Tails
print("Coins flipped head: ",coins)
print("Number of coins flipped head: ",sum(coins))
```

```python
# Every time we run, we randomly get a different number of heads
# So the average results will also be random
# However, according to the Central Limit Theorem, the average results approach the true mean the more times I run

# Now let's apply the Central Limit Theorem to see if the results allow us to identify an unfair coin
# Coin-flipping hypothesis testing
n_coins = 10
n_trials = 100
```

```python
# Flip 1st coin
p = 0.5 # fair coin
n_heads = []
for i in range(n_trials): # repreat 100 trials
  coins = (random.random(n_coins)) < p # at each trial, throw 10 coins
  n_heads.append(sum(coins))

# Calculate mean and standard deviation of all results
mean_1, sd_1 = (np.mean(n_heads), np.std(n_heads))
print("Mean: ",mean_1, "; StdDev: ", sd_1)

# Calculate the 95% confidence interval = mean +- 1.95 x standard deviation
print("95% confidence interval for mean: [", round(mean_1 - 1.95*sd_1/np.sqrt(n_trials), 2), ",", round(mean_1 + 1.95*sd_1/np.sqrt(n_trials),2),"]")

# Plot the histogram of the 100 trials. The average is close to 5
plt.hist(n_heads, np.max(n_heads) - np.min(n_heads) +1, density=False, facecolor='g',alpha=0.75)
plt.xlabel('Number of heads')
plt.ylabel('Number of trials')
plt.grid(True)
plt.show()
```

```python
# Flip 2nd coin
p = 0.4 # unfair coin - 40% chance of Heads
n_heads = []
for i in range(n_trials): # repreat 100 trials
  coins = (random.random(n_coins)) < p # at each trial, throw 10 coins
  n_heads.append(sum(coins))

# Calculate mean and standard deviation of all results
mean_2, sd_2 = (np.mean(n_heads), np.std(n_heads))
print("Mean: ",mean_2, "; StdDev: ", sd_2)

# Calculate the 95% confidence interval = mean +- 1.95 x standard deviation
print("95% confidence interval for mean: [", round(mean_2 - 1.95*sd_2/np.sqrt(n_trials), 2), ",", round(mean_2 + 1.95*sd_2/np.sqrt(n_trials),2),"]")

# Plot the histogram of the 100 trials. The average is close to 4
plt.hist(n_heads, np.max(n_heads) - np.min(n_heads) +1, density=False, facecolor='g',alpha=0.75)
plt.xlabel('Number of heads')
plt.ylabel('Number of trials')
plt.grid(True)
plt.show()
```

## 3.2. Hypothesis testing

```python
# normal distribution:

# cdf(t) function indicates the probability(random sample < t)
print(norm.cdf(0))
print(norm.cdf(1.64))
print(norm.cdf(1.96))

# ppf(x) function indicates the probability(random sample < t)
print(norm.ppf(0.5))
print(norm.ppf(0.95))
print(norm.ppf(0.975)) # This means that 95% of the samples may fall within 1.95 standard deviations from the mean
```

```python
# HYPOTHESIS TESTING

# First, we apply to the Coin Flipping Example
# First calculate T-value : How distant data is from the hypothesis
t = (mean_1 - mean_2) * np.sqrt(n_trials) / np.sqrt(sd_1**2 + sd_2**2)
print("t-value: ", t)

# Then calculate the p-value = change the diference to the Null Hypothesis was just a coincidence
# the sample mean follows a normal distribution, because of the Central Limit Theorem
p = 1 - norm.cdf(t)
print("p-value: ", p)
```

## 3.3. Linear model and regression

```python
# The linear model assumes y as a linear function of y.
# If there is no random component, "x x y" it will be just a straight line
n = 20
a = 2
b = 3
x = np.array(range(-n,n))
y = a*x +b
plt.scatter(x,y)
plt.plot([-n,n],[a*(-n)+b,a*(n)+b],'k-')
plt.show
```

```python
# The linear model assumes a random component:
# y = ax + b + error
# Here, we assume the error to be normally distributed with mean 0 and some standard deviation
# What is the effect of increasing or decreasing the standard deviation?

std_dev = 50
random_draws = np.random.normal(0, std_dev, 2*n)
y = a*x + b + random_draws # y = ax + b + error

# Plot the samples
plt.scatter(x,y)
plt.plot([-n,n],[a*(-n)+b,a*(n)+b],'k-')
plt.show
```

```python
# The Linear Regression tries to identify, from the data, the coefficients a and b of the model

# As the model assumes some random error in the data, then the coefficients obtained will also be random variables
# (different from the true coefficients, that are not random but are unknown)
# The model returns the estimated coefficents and the estimation standard error. 
# From the estimation and the standard errors, we can calculate the Confidence Interval (usially with 95% certainty)

# The model explains a fraction of the variation in the data. The fraction of the variation explained by the model is called R-Square

X = sm.add_constant(x)
est = sm.OLS(y, X) # or sm.RLM(y, x) for robust standard deviation, due to heteroscedasticity
est2 = est.fit()
print(est2.summary())
```

```python
# These are the coefficients infered by our model. Compare with the true coefficients, from our data generation process
print(est2.params)
b_estimate = est2.params[0]
a_estimate = est2.params[1]

# In black, the line by the true coefficients
# In red, the line "ax+b" of the coefficients obtained by our linear regression
plt.scatter(x,y)
plt.plot([-n,n],[a*(-n)+b, a*(n)+b],'k-')
plt.plot([-n,n],[a_estimate*(-n)+b_estimate, a_estimate*(n)+b_estimate],'r-')
plt.show
```

```python
def distance(x1, x2):
  #print(x1)
  #print(x2)
  if len(x1) != len(x2):
    print('x1 and x2 should have same distance')
    return 0
  return np.sqrt(sum([(x1[i]-x2[i])**2 for i in range(len(x1))]))

def closest_cluster(x, clusters):
  distances = []
  #print(x)
  #print(clusters)
  for cluster in clusters:
    distances.append(distance(list(x), cluster))
  return distances.index(min(distances))

def plot_clusters(data,clusters, iteration):
  colors = ['black','red','green','blue','yellow']
  # Plot the cluster and the actual formation for each data point as a function of depth
  for i in range(len(data)):
    plt.scatter(data[list(data)[0]][i], data[list(data)[1]][i], c=colors[data['cluster'][i]], marker=".", s = 3)
  for i in range(len(clusters)):
    plt.scatter(clusters[i][0], clusters[i][1], c=colors[i], marker="o", s = 30)
  plt.xlabel(list(data)[0])
  plt.title('Iteration #'+str(iteration))
  plt.ylabel(list(data)[1])
  plt.show()

def cluster(data, number_of_clusters, number_of_iterations):
  data = data.dropna()
  original_data = data.copy()
  k = len(list(data))
  n = len(data)
  loss = []
  data['cluster'] = 0
  data['distance'] = 10

  cluster_points = np.random.choice(n, number_of_clusters, replace=False)
  clusters = [list(original_data.iloc[centroid]) for centroid in cluster_points] # create random clusters

  for itert in range(number_of_iterations):
    # Plot locations and alocations
    plot_clusters(data,clusters, itert+1)

    # Assign points to closest cluster
    for i in range(n):
      cluster_i = closest_cluster(list(original_data.iloc[i]), clusters)
      data.iloc[i,k] = cluster_i #['cluster']
      data.iloc[i,k+1] = distance(list(original_data.iloc[i]), clusters[cluster_i]) #['distance']

    # Update cluster locations
    for i in range(number_of_clusters):
      for j in range(k):
        mean = np.mean(data.loc[data['cluster'] == i, list(data)[j]])
        clusters[i][j] = mean
    
    loss.append(sum(data['distance']))

  plt.plot(range(number_of_iterations), loss)
  plt.xlabel('Iteration')
  plt.ylabel('Sum of distances')
  plt.show()

  return loss[number_of_iterations-1]
    
```
